**Plotando preços de fechamento da ação AAPL(Apple)**

```
# import chart_studio.plotly as py
import plotly.graph_objs as go

data = [go.Scatter(
          x = df.Date,
          y = df['Close'])]

py.iplot(data)
```

Customizando Faixs de Eixos (Range)

```
Close = go.Scatter(
                x = df.Date,
                y = df.Close,
                name = "Petr4 High",
                line = dict (color = '#17BECF'),
                opacity = 0.8)

data = [Close]

layout = dict (
    title = "Faixa de Datas Customizada",
    title_x = 0.5,
    xaxis = dict(
        range = ['2019-01-01', '2020-09-24'])
)

fig = dict (data = data, layout = layout)
py.iplot(fig, filename = "Definir intervalo manualmente")
```

**Rangeslider e Rangeselector**

```
Close = go.Scatter(
                x = df.Date,
                y = df.Close,
                name = "Petr4 High",
                line = dict (color = '#17BECF'),
                opacity = 0.8)

data = [Close]

layout = dict (
    title = "Série com Rangeslider e Botões",
    title_x = 0.5,
    xaxis = dict (
        rangeselector=dict (
            buttons = list ([
                dict (count= 1,
                      label = '1 mês',
                      step = 'month',
                      stepmode = 'backward'),
                dict (count=6,
                      label = '6 meses',
                      step = 'month',
                      stepmode = 'backward'),
                dict (step = 'all')
            ])
        ),
        rangeslider = dict (
            visible = True
        ),
        type = 'date'
    )
)

fig = dict (data = data, layout = layout)
py.iplot(fig)
```

Trabalhando com Candlesticks

```
trace = go.Candlestick (x=df['Date'],
                        open = df['Open'],
                        high=df['High'],
                        low=df['Low'],
                        close=df['Close'])

data = [trace]

layout = {
    'title': 'Petr4 - Preços no Período',
    'title_x': 0.5,
    'yaxis': {'title':'Petr4 - Preço'},
    'annotations': [{
        'x': '2020-02-27',
        'y': 35,
        'xref': 'x',
        'yref': 'y',
        'showarrow': True,
        'font': dict (
            family="Corrier New, monospace",
            size=12
        ),
        'text': 'Crise da Covid-19 no Brasil',
        'align': "center",
        'arrowhead': 2,
        'arrowsize': 1,
        'arrowwidth': 2,
        'bordercolor': "#c7c7c7",
        'borderwidth': 2,
        'borderpad': 4
    }]
}

#py.iplot(data, filename=  'Simple_Candlestick')
fig = dict (data=data, layout=layout)
py.iplot(fig)
```

**Trabalhando com Médias**

```
# Média Simples dos últimos 07 dias
df['Média_07_dias'] = df.Close.rolling(window=7).mean()

# Média Simples dos últimos 15 dias
df['Média_15_dias'] = df.Close.rolling(window=15).mean()

# Média Simples dos últimos 30 dias
df['Média_30_dias'] = df.Close.rolling(window=30).mean()
```

```
# Visualizando colunas criadas
df.head(31)
```

Plotar as médias e os preços

```
close = go.Scatter(
    x=df.Date,
    y=df.Close,
    name="Petr4 Close",
    line=dict (color = '#330000'),
    opacity=0.8)

Média_07_dias = go.Scatter(
    x=df.Date,
    y=df['Média_07_dias'],
    name="Média últimos 07 dias",
    line=dict (color = '#FF00FF'),
    opacity=0.8)

Média_15_dias = go.Scatter(
    x=df.Date,
    y=df['Média_15_dias'],
    name="Média últimos 15 dias",
    line=dict (color = '#FFD500'),
    opacity=0.8)

Média_30_dias = go.Scatter(
    x=df.Date,
    y=df['Média_30_dias'],
    name="Média últimos 30 dias",
    line=dict (color = '#1c10a3'),
    opacity=0.8)

data = [close, Média_07_dias, Média_15_dias, Média_30_dias]
py.iplot(data)
```
